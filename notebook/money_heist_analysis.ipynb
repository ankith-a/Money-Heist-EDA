```python
import pandas as pd
import plotly.express as px

df4 = pd.DataFrame([
    dict(task='Season1', Start='02-May-2017', Finish='20-Jun-2017', completion_act=25),
    dict(task='Season2', Start='06-Apr-2018', Finish='06-Apr-2018', completion_act=25),
    dict(task='Season3', Start='19-Jul-2019', Finish='19-Jul-2019', completion_act=25),
    dict(task='Season4', Start='03-Apr-2020', Finish='03-Apr-2020', completion_act=25),
    dict(task='Season5', Start='03-Sep-2021', Finish='03-Sep-2021', completion_act=25)
])

# Convert to datetime
df4['Start'] = pd.to_datetime(df4['Start'], format='%d-%b-%Y')
df4['Finish'] = pd.to_datetime(df4['Finish'], format='%d-%b-%Y')

fig = px.timeline(
    df4,
    x_start='Start',
    x_end='Finish',
    y='task',
    color='task'
)

fig.update_yaxes(autorange='reversed')
fig.show()
```
